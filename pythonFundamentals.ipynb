## **Variables in Python | Dynamic Typing | Dynamic Binding (Lecture - 03,04)**

### Static vs Dynamic Typing

- **Dynamic Typing (Python)** → Type is decided at runtime and can change  
  Example:  
  x = 10  
  x = "hello"   # type changed from int to str  

- **Static Typing** → Type is fixed before execution and cannot change  
  Example (concept):  
  int x = 10  


### Static vs Dynamic Binding

- **Static Binding** → Type/value is fixed at compile-time  

- **Dynamic Binding** → Type/value is decided at runtime and can change during execution

## **Operators (Lecture - 09)**

- ### Arithmetic Operators : `+`, `-`, `*`, `/`, `%`, `**`, `//`
  Used for mathematical operations.

- ### Comparison Operators : `==`, `!=`, `>`, `<`, `>=`, `<=`
  Used to compare values.  

- ### Logical Operators : `and`, `or`, `not`
  Used to combine conditional statements.  

- ### Bitwise Operators : `&`, `|`, `^`, `~`, `<<`, `>>`
  Operate on binary representations.  

- ### Assignment Operators : `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`
  Used to assign values to variables.  

- ### Identity Operators : `is`, `is not`
  Used to compare memory locations of objects.  

- ### Membership Operators : `in`, `not in`
  Used to test if a value is in a sequence.  

```
# Arithmetic Operators

x = 5
y = 2
print("x + y =", x + y)
print("x - y =", x - y)
print("x / y =", x / y)
print("x % y =", x % y)
print("x * y =", x * y)
print("x ** y =", x ** y)
print("x // y =", x // y)
```

```
# Comparison Operators

print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
print(x == y)
print(x != y)
```

```
# Logical Operators

x = True
y = False

print(x or y)   # Returns True if at least one value is True
print(x and y)  # Returns True only if both values are True
print(not x)    # Reverses the value of x (True → False)
print(not y)    # Reverses the value of y (False → True)
```

```
# Bitwise Operators

x = 2  # binary: 010
y = 3  # binary: 011

print("x =", x, "->", bin(x))
print("y =", y, "->", bin(y))
print()

print("x & y (AND): ", x & y, "->", bin(x & y))
# AND: 1 only if both bits are 1

print("x | y (OR): ", x | y, "->", bin(x | y))
# OR: 1 if at least one bit is 1

print("x ^ y (XOR): ", x ^ y, "->", bin(x ^ y))
# XOR: 1 if bits are different

print("x >> 2 (Right Shift): ", x >> 2, "->", bin(x >> 2))
# Right Shift: shifts bits to the right (divide by 2^n)

print("y << 3 (Left Shift): ", y << 3, "->", bin(y << 3))
# Left Shift: shifts bits to the left (multiply by 2^n)

print("~x (NOT): ", ~x, "->", bin(~x))
# NOT: inverts all bits (uses 2's complement)
```

```
# Assignment Operators

a = 3
print("Initial value of a:", a)
print()

a += 3   # a = a + 3
print("After a += 3  ->", a)

a -= 3   # a = a - 3
print("After a -= 3  ->", a)

a *= 3   # a = a * 3
print("After a *= 3  ->", a)

a /= 3   # a = a / 3
print("After a /= 3  ->", a)

# Increment / Decrement Operators
# --a, a--, ++a, a++ are NOT supported in Python
# Instead, use:
a += 1   # increment
a -= 1   # decrement
```

```
# Identity Operators

# Case 1: Integers
a = 3
b = 3
print("a =", a, ", b =", b)
print("a is b ->", a is b)
# True: Small integers may share the same memory location
print()

# Case 2: Strings
a = "Hello"
b = "Hello"
print("a =", a, ", b =", b)
print("a is b ->", a is b)
# True: Python may reuse (intern) strings
print()

# Case 3: Lists
a = [1, 2, 3]
b = [1, 2, 3]
print("a =", a, ", b =", b)
print("a is b ->", a is b)
# False: Lists are different objects in memory
```

```
# Membership Operators

# Case 1: String
x = "Delhi"
print("String:", x)
print("D" in x)       # True: 'D' is present
print("D" not in x)   # False: 'D' is present
print()

# Case 2: List
x = [1, 2, 3]
print("List:", x)
print("1 in x  ->", 1 in x)   # True: 1 is present
print("5 in x  ->", 5 in x)   # False: 5 is not present
print()

# Case 3: Tuple
x = (1, 2, 3)
print("Tuple:", x)
print("1 in x  ->", 1 in x)   # True: 1 is present
print("5 in x  ->", 5 in x)   # False: 5 is not present
```
